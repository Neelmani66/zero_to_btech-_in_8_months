Data Types :

1. Integers - int - 3,33,333 

2. Floating points - float - 3.0 , 33.0 , 333.33

3. Strings - str - "hello" , 'sammy'

4. Lists (ordered sequence of objects) - list - [10 , 'hello' , 200.3]

5. Dictionaries (unordered key : value pairs) - dict - {"mykey" : "value" , "name" : "frankie"}

6. Tuples (ordered immutable sequence of objects) - tup - (10, "hello", 200.3)

7. Sets (unordered collection of unique object) - set - ("a","b")

8. Boolean - bool - true/false
